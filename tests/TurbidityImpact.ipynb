```python
from ultralytics import YOLO
import os


dataset_path = r"C:\Users\Test\Documents\ThesisVideos\VideoDatasetSegmented"

folders = ["NTU85_105","NTU110_130","NTU130_150","NTU158_176","NTU183_198"]

for folder in folders:
    
    val_path = os.path.join(dataset_path, folder, "valid\data.yaml")

    metrics = YOLOv8.val(
        data=val_path,
        split="val",
        imgsz=640,
        batch=16,
        device=0,
        save=False,
        val=True
    )

    print(f"\nResults for {folder}")
    print("mAP50:", metrics.box.map50)
    print("Precision:", metrics.box.mp)
    print("Recall:", metrics.box.mr)
```
